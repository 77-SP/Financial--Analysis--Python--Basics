# IPL 2026 — Which Team Got the Best ROI?
---
This project analyzes all 10 IPL 2026 teams and calculates their Return on Investment.

We compare each team's squad spend against their league stage wins to find out who spent smartest.

## Step 1 : Import Libraries
---
### What are we importing?

We import two libraries:
- **matplotlib** is used to create charts and visualizations.
- **numpy** is used for mathematical calculations like trend lines.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

print('Libraries imported successfully!')
```

## Step 2 : Load IPL 2026 Data
---
### What data are we using?

We manually enter the real IPL 2026 data for all 10 teams:
- **teams** — names of all 10 IPL franchises.
- **spend** — total squad cost in crore rupees (source: cricjosh.in).
- **wins** — total league stage wins (source: ESPNcricinfo).

### What will the output show?

A simple table showing each team, how much they spent, and how many games they won.

```python
# All 10 IPL 2026 teams
teams = ['RCB', 'GT', 'MI', 'RR', 'DC', 'PBKS', 'SRH', 'LSG', 'KKR', 'CSK']

# Total squad spend in Crore (source: cricjosh.in/ipl-2026-salary)
spend = [88.7, 84.8, 90.2, 86.3, 86.9, 92.1, 91.5, 89.6, 87.1, 85.4]

# League stage wins (source: ESPNcricinfo IPL 2026 Points Table)
wins = [9, 9, 7, 7, 6, 6, 6, 6, 5, 6]

# Print summary table
print(f"{'Team':<8} {'Spend (Cr)':>12} {'Wins':>6}")
print('-' * 28)
for t, s, w in zip(teams, spend, wins):
    print(f"{t:<8} {s:>12.1f} {w:>6}")
```

## Step 3 : Calculate ROI
---
### What is ROI?

ROI stands for Return on Investment. In simple terms, it tells us how much output we got for every rupee we invested.

For IPL teams we calculate it as:

**ROI = (Wins / Squad Spend) x 10**

This gives us wins per Rs 10 Crore spent — making it easy to compare all teams fairly.

### What will the output show?

A ranked list of all 10 teams from highest ROI to lowest, showing who got the most value from their spending.

```python
# Calculate ROI for each team
roi = [round((w / s) * 10, 3) for w, s in zip(wins, spend)]

# Sort by ROI from highest to lowest
sorted_data = sorted(zip(roi, teams, spend, wins), reverse=True)
roi_s, teams_s, spend_s, wins_s = zip(*sorted_data)

# Print ROI ranking
print('IPL 2026 — ROI RANKINGS')
print(f"{'Rank':<6} {'Team':<8} {'ROI':>10} {'Spend (Cr)':>12} {'Wins':>6}")
print('-' * 46)
for i, (r, t, s, w) in enumerate(zip(roi_s, teams_s, spend_s, wins_s), 1):
    tag = '  <-- CHAMPIONS' if t == 'RCB' else ''
    print(f"{i:<6} {t:<8} {r:>10.3f} {s:>12.1f} {w:>6}{tag}")
```

## Step 4 : Visualize the Results
---
### What charts are we building?

We build two charts side by side:
- **Chart 1 — ROI Ranking** shows which team got the most wins per crore spent, ranked from best to worst.
- **Chart 2 — Spend vs Wins** shows whether spending more money actually led to more wins.

### What will the output show?

Two dark themed charts with all 10 teams in their official colors. RCB is highlighted in gold as the IPL 2026 Champions.

```python
# Official IPL team colors
team_colors = {
    'RCB': '#D41B2C', 'GT': '#1C4E9D', 'MI': '#005DA0',
    'RR': '#EA1A85',  'DC': '#17479E', 'PBKS': '#DCAB38',
    'SRH': '#F7A721', 'LSG': '#4A90D9', 'KKR': '#3B215E', 'CSK': '#F9CD05',
}
bar_colors = [team_colors[t] for t in teams_s]

fig, axes = plt.subplots(1, 2, figsize=(16, 7), facecolor='#0D1117')

# --- Chart 1: ROI Bar Chart ---
ax1 = axes[0]
ax1.set_facecolor('#0D1117')
bars = ax1.barh(list(teams_s), list(roi_s), color=bar_colors, edgecolor='white', linewidth=0.4, height=0.6)

for bar, team in zip(bars, teams_s):
    if team == 'RCB':
        bar.set_edgecolor('gold')
        bar.set_linewidth(2.5)

for i, (val, team) in enumerate(zip(roi_s, teams_s)):
    label = f'{val:.3f}  CHAMPIONS' if team == 'RCB' else f'{val:.3f}'
    ax1.text(val + 0.0005, i, label, va='center', ha='left',
             color='gold' if team == 'RCB' else 'white', fontsize=9, fontweight='bold')

ax1.set_xlabel('Wins per Rs 10 Crore Spent', color='#AAAAAA', fontsize=11)
ax1.set_title('IPL 2026 ROI Ranking\n(Wins per Rs 10 Cr of Squad Spend)', color='white', fontsize=13, fontweight='bold', pad=12)
ax1.tick_params(colors='white', labelsize=10)
for spine in ['top', 'right']: ax1.spines[spine].set_visible(False)
for spine in ['bottom', 'left']: ax1.spines[spine].set_color('#444')
ax1.set_xlim(0, max(roi_s) * 1.3)

# --- Chart 2: Spend vs Wins Scatter ---
ax2 = axes[1]
ax2.set_facecolor('#0D1117')

for team, s, w in zip(teams, spend, wins):
    ax2.scatter(s, w, color=team_colors[team], s=220, edgecolors='white', linewidths=0.8, zorder=3)
    label = f'{team} (Champions)' if team == 'RCB' else team
    ax2.annotate(label, (s, w), textcoords='offset points', xytext=(6, 4),
                 color='gold' if team == 'RCB' else '#CCCCCC',
                 fontsize=8.5, fontweight='bold' if team == 'RCB' else 'normal')

z = np.polyfit(spend, wins, 1)
p = np.poly1d(z)
x_line = np.linspace(min(spend)-1, max(spend)+1, 100)
ax2.plot(x_line, p(x_line), '--', color='#888', linewidth=1.2)

ax2.set_xlabel('Total Squad Spend (Rs Crore)', color='#AAAAAA', fontsize=11)
ax2.set_ylabel('League Stage Wins', color='#AAAAAA', fontsize=11)
ax2.set_title('IPL 2026 - Spend vs Performance\n(Does spending more = winning more?)', color='white', fontsize=13, fontweight='bold', pad=12)
ax2.tick_params(colors='white', labelsize=10)
for spine in ['top', 'right']: ax2.spines[spine].set_visible(False)
for spine in ['bottom', 'left']: ax2.spines[spine].set_color('#444')
ax2.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

fig.text(0.5, 0.01, 'Finance With Samya | Episode 44 | Data: cricjosh.in & ESPNcricinfo',
         ha='center', color='#888888', fontsize=8.5)
plt.suptitle('Which IPL 2026 Team Got the Best Return on Investment?',
             color='white', fontsize=15, fontweight='bold', y=1.02)

plt.tight_layout()
plt.savefig('ipl2026_roi_chart.png', dpi=150, bbox_inches='tight', facecolor='#0D1117')
plt.show()
print('Charts displayed successfully!')
```
