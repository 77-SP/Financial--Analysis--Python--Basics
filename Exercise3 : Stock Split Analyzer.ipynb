# **Exercise 3 : Stock Split Ananlyzer**

This is an excellent Python-for-finance project that demonstrates how a stock split changes the number of shares and price per share while keeping the total investment value unchanged.

 I have build a program that calculates:


*   Share Before Split
*   Share After Split


*   Price Before Split
*   Price After Split

*  Total Value Before Split
*  Total Value After Split



# **Finance Concept**

---
In a stock split:


*   The number of shares increases.
*   The price per share decreases proportionally.

*   The Total Investment value remains the same.



# **Example:2-for-1 Stock Split**

---





*   Before:  10 shares × ₹1,000 = ₹10,000



*   After:  20 shares × ₹500 = ₹10,000








# **Company Information**

---



```python
company="Apple"
shares_before= 10
price_before= 1000
split_ratio=2 #2-for-1 stock split
```

#**Calculate Shares after Split**

---



```python
share_after=shares_before*split_ratio
print("Shares after split: ",share_after)
```

# **Calculate Price after Split**

---



```python
price_after=price_before/split_ratio
print("Price After Split :₹",price_after)
```

# **Calculate Total Value Before Split**

---



```python
value_before=shares_before*price_before
print("Value Before Split :₹",value_before)
```

# **Calculate Total Value After Split**

```python
value_after=share_after*price_after
print("Value After Split :₹",value_after)
```

# **Final  Summary**

---



```python
print("\n--- Stock Split Summary ---")
print("Company:", company)
print("Split Ratio:", f"{split_ratio}-for-1")
print("Shares Before Split:", shares_before)
print("Shares After Split:",share_after)
print("Price Before Split: ₹", price_before)
print("Price After Split: ₹", price_after)
print("Value Before Split: ₹", value_before)
print("Value After Split: ₹", value_after)
```

# **Summary Table**

---



```python
import pandas as pd

summary = pd.DataFrame({
    'Metric': [
        'Shares Before Split',
        'Shares After Split',
        'Price Before Split',
        'Price After Split',
        'Value Before Split',
        'Value After Split'
    ],
    'Value': [
        shares_before,
        share_after,
        f'₹{price_before:,.2f}',
        f'₹{price_after:,.2f}',
        f'₹{value_before:,.2f}',
        f'₹{value_after:,.2f}'
    ]
})
summary
```

# **Interpretation**

---


A stock split increases the number of shares and reduces the price per share proportionally.

However, the total investment value remains unchanged.
