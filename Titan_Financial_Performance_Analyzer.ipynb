#**PHASE 1 : Core Financial Analysis**

---



##**Step 1** **:** **Import Libraries**

---



```python
import pandas as pd
import matplotlib.pyplot as plt
```

##**Step 2 : Load Financial Data**

---



In this step, the Excel dataset containing Titan's financial information is imported into Pandas for analysis.

```python
df = pd.read_excel("Titan's Financial Data.xlsx", header=1)

df
```

## **Step 3 : Explore the Dataset**

---



```python
df.head()
```

##**Step 4: Financial Ratio Analysis**

---



Key financial ratios are calculated to assess profitability, efficiency, and leverage.

```python
df["Net Profit Margin (%)"]=(df["Net Income"]/df["Revenue"])*100

df
```

**Formula :**

---
Net Profit Margin = Net Income/Revenue *100


## **Step 5: Calculate ROA**

---



```python
df["Return on Assets(%)"] = (df["Net Income"]/df["Total Assets"])*100

df
```

**Formula:**

---



Return on Asset (ROA) = Net Income / Total Assets *100

## **Step 6: Calculate ROE**

---



```python
df[("Return on Equity (%)")] = (df["Net Income"]/df["Total Equity"])*100

df
```

**Formula:**

---
Return on Equity (ROE) = Net Income / Total Equity * 100


##**Step 7: Calculate Debt-to-Equity**

---



```python
df[("Debt-to-Equity")] = (df["Borrowings"]/df["Total Equity"])

df
```

**Formula:**

---
Debt-to-Equity = Borrowings / Total Equity


##**Step 8: Round the Results**

```python
df.round(2)
```

# **PHASE 2: Growth Analysis**

---



Year-over-year growth metrics help evaluate Titan's business expansion and profitability trends.

##**Step 9: Revenue Growth %**

```python
df["Revenue Growth (%)"] = (
    df["Revenue"].pct_change()
) * 100

df.round(2)
```

##**Step 10: Net Income Growth %**

```python
df["Net Income Growth (%)"] = (
    df["Net Income"].pct_change()
) * 100

df.round(2)
```

# **PHASE 3: Visualization**

---



Visual representations provide a clearer understanding of financial trends over time.

##**Step 11: Import Matplotlib**

```python
import matplotlib.pyplot as plt
```

##**Step 12: Revenue Trend Chart**

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["Year"],
    df["Revenue"],
    marker="o"
)

plt.title("Titan Revenue Trend")
plt.xlabel("Year")
plt.ylabel("Revenue (₹ Crore)")

plt.grid(True)
plt.savefig("revenue_trend.png")
plt.show()
```

##**Step 13 : Net Income Trend Chart**

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["Year"],
    df["Net Income"],
    marker="o"
)

plt.title("Titan Net Income Trend")
plt.xlabel("Year")
plt.ylabel("Net Income (₹ Crore)")

plt.grid(True)
plt.savefig("net_income_trend.png")
plt.show()
```

##**Step 14: ROE Trend Chart**

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["Year"],
    df["Return on Equity (%)"],
    marker="o"
)

plt.title("Titan ROE Trend")
plt.xlabel("Year")
plt.ylabel("ROE (%)")
plt.grid(True)
plt.show()
```

##**Step 15: Final Dataset**

```python
df.round(2)
```

#**PHASE 4: Automated Fianancial Analysis Report**

---



Python generates a financial summary based on the latest available data and predefined performance benchmarks.

---



##**Step 16: Create a Report Header**






```python

print("=" * 50)
print("TITAN COMPANY FINANCIAL ANALYSIS REPORT")
print("=" * 50)
```

##**Step 17: Latest Year Sanpshot**

```python
latest = df.iloc[-1]

print(f"Year: {latest['Year']}")
print(f"Revenue: ₹{latest['Revenue']:,.0f} Cr")
print(f"Net Income: ₹{latest['Net Income']:,.0f} Cr")
print(f"Return on Equity: {latest['Return on Equity (%)']:.2f}%")
print(f"Return on Assets: {latest['Return on Assets(%)']:.2f}%")
print(f"Net Profit Margin: {latest['Net Profit Margin (%)']:.2f}%")
print(f"Debt-to-Equity: {latest['Debt-to-Equity']:.2f}")
```

## **Step 18: Revenue Growth Insight**

```python
latest_growth = df["Revenue Growth (%)"].iloc[-1]

if latest_growth > 10:
    print("✓ Revenue growth is strong.")
elif latest_growth > 0:
    print("✓ Revenue is growing steadily.")
else:
    print("⚠ Revenue has declined.")
```

## **Step 19: Profitabilty Insight**

```python
latest_margin = df["Net Profit Margin (%)"].iloc[-1]

if latest_margin > 10:
    print("✓ Profitability is strong.")
elif latest_margin > 5:
    print("✓ Profitability is healthy.")
else:
    print("⚠ Profitability is relatively low.")
```

##**Step 20: Return on Equity Insight**

```python
latest_roe = df["Return on Equity (%)"].iloc[-1]

if latest_roe > 20:
    print("✓ Excellent shareholder returns.")
elif latest_roe > 15:
    print("✓ Good shareholder returns.")
else:
    print("⚠ ROE could be improved.")
```

## **Step 21: Debt Insight**

```python
latest_de = df["Debt-to-Equity"].iloc[-1]

if latest_de < 0.5:
    print("✓ Conservative debt levels.")
elif latest_de < 1:
    print("✓ Moderate leverage.")
else:
    print("⚠ High dependence on debt financing.")
```

##**Step 22: Overall Conclusion**

```python
print("\nOVERALL CONCLUSION")

if latest_roe > 20 and latest_growth > 10:
    print("Titan demonstrates strong growth and profitability.")
elif latest_growth > 0:
    print("Titan shows stable financial performance.")
else:
    print("Titan may require closer financial review.")
```
