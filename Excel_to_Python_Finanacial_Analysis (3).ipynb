# **Excel to Python Financial Analysis**

---



This project demonstrates how financial data stored in Excel can be analyzed using Python.

The analysis includes key financial ratios and visualizations for company comparison.

# **Step 1** : **Import Libraries**

---


#**Import & Load Data from Excel File**

---



```python
import pandas as pd

df = pd.read_excel("Company_Data (1).xlsx")

print(df)
```

# **Calculate Financial Ratios**

```python
df["Profit Margin (%)"] = (df["Net Income"] / df["Revenue"]) * 100

df["Current Ratio"] = (
    df["Current Assets"] / df["Current Liabilities"]
)

df["Debt-to-Equity Ratio"] = (
    df["Total Debt"] / df["Shareholder Equity"]
)
print(df)
```

#**Revenue Comparison Chart**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,5))

plt.bar(df["Company"], df["Revenue"])
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Revenue Comparison")

plt.show()
```

#**Profit Margin Comparison Chart**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,5))

plt.bar(df["Company"], df["Profit Margin (%)"])
plt.xlabel("Company")
plt.ylabel("Profit Margin (%)")
plt.title("Profit Margin Comparison")
plt.show()
```

# **Debt-to-Equity Comparison**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,5))

plt.bar(df["Company"], df["Debt-to-Equity Ratio"])
plt.xlabel("Company")
plt.ylabel("Debt-to-Equity")
plt.title("Debt-to-Equity Comparison")
plt.show()
```
