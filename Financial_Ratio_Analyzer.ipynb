# **Financial Ratio Analyzer**

This cell installs the `yfinance` library, which is used to fetch financial data from Yahoo Finance.

## **Step 2: Import Libraries**

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

This cell imports the necessary Python libraries: `yfinance` for financial data, `pandas` for data manipulation, and `matplotlib.pyplot` for plotting.

## **Step 3: Define Stocks**

```python
stocks = ["AAPL", "MSFT", "TCS.NS"]
```

This cell defines a list of stock tickers for which we want to retrieve financial data. 'AAPL' is Apple, 'MSFT' is Microsoft, and 'TCS.NS' is Tata Consultancy Services (from the Indian market).

##**Step 4: Initialize Data List**

```python
data = []
```

This cell initializes an empty list called `data`, which will be used to store the fetched financial metrics for each stock.

## ** Step 5: Fetch Stock Data**

```python
for stock in stocks:
    company = yf.Ticker(stock)
    info = company.info

    data.append({
        "Company": stock,
        "P/E Ratio": info.get("trailingPE"),
        "ROE": info.get("returnOnEquity"),
        "Profit Margin": info.get("profitMargins")
    })
```

This loop iterates through each stock ticker, fetches its company information using `yf.Ticker(stock).info`, and then extracts the P/E Ratio, Return on Equity (ROE), and Profit Margin. These metrics are then appended as a dictionary to the `data` list.

## Create DataFrame

```python
df = pd.DataFrame(data)
```

This cell converts the list of dictionaries (`data`) into a pandas DataFrame, which is a tabular data structure that is easier to work with for analysis.

## Display DataFrame

```python
print(df)
```

This cell prints the entire DataFrame to display the fetched financial data for all the selected companies.

## P/E Ratio Comparison Plot

```python
df.plot(
    x="Company",
    y="P/E Ratio",
    kind="bar"
)

plt.title("P/E Ratio Comparison")
plt.ylabel("P/E Ratio")

plt.show()
```

This cell generates a bar plot to visually compare the P/E Ratios of the selected companies. The 'Company' column is used for the x-axis, and 'P/E Ratio' for the y-axis.

## ROE Comparison Plot

```python
df.plot(
    x="Company",
    y="ROE",
    kind="bar"
)

plt.title("ROE Comparison")

plt.show()
```

This cell generates another bar plot, this time to compare the Return on Equity (ROE) of the selected companies. The 'Company' column is on the x-axis, and 'ROE' is on the y-axis.
