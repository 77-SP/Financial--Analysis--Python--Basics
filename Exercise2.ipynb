##**Exercise 2**: Return on Investment with Quantity

**Objective:**

The objective of this exercise is to extend stock return analysis by including the number of shares (quantity) and calculating the actual profit or loss in monetary terms, along with percentage return.

---



**Profit, Return and Quantity Calculation**

**Scenario:**

---



**EXAMPLE 1:**

---



Suppose an investor buys a stock at ₹200 and later sells it at ₹250.

The investor purchases **10 shares**.


We want to calculate:


*   Total Profit or Loss
*   Percentage Return
*   Actual Profit based on Quantity



```python
buy_price= 200
sell_price= 250
quantity= 10

profit= (sell_price-buy_price)
total_profit= profit*quantity
percentage_return= (profit/buy_price)*100

print("Profit per share:",profit)
print("Total profit:",total_profit)
print("Return(%)",percentage_return)

if total_profit > 0:
  print("profit")
else:
  print("loss")
```



---



**EXAMPLE 2:**

Suppose an investor buys a stock at ₹500 and sells it at ₹450 with **5 shares**.

```python
buy_price=500
sell_price=450
quantity=5

profit= (sell_price-buy_price)
total_profit= profit*quantity
percentage_return= (profit/buy_price)*100

print("Profit per share:",profit)
print("Total profit:",total_profit)
print("Return(%)",percentage_return)

if total_profit > 0:
  print("profit")
else:
  print("loss")
```
