# 1: Profit and Return Calculation

## Scenario
Suppose an investor buys a stock at ₹200 and later sells it at ₹250.

We want to calculate:
- How much profit or loss is made
- What is the percentage return on investment

```
buy_price = 200
sell_price = 250

profit = sell_price - buy_price
return_percent = (profit / buy_price) * 100

print("Profit:", profit)
print("Return (%):", return_percent)
```
